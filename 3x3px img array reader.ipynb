```python
# Create a figure and axis
fig, ax = plt.subplots()

# Display the image
ax.imshow(rgb_image)

# Add pixel values as text
for i in range(rgb_image.shape[0]):
    for j in range(rgb_image.shape[1]):
        r, g, b = rgb_image[i, j]
        ax.text(j, i, f"{r},{g},{b}", color='white', ha='center', va='center',
                path_effects=[pe.Stroke(linewidth=1.5, foreground='black'), pe.Normal()]
               )

# Hide ticks and labels
ax.set_xticks([])
ax.set_yticks([])

# Show the plot
plt.show()
```
